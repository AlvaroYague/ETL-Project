```python
import pandas as pd
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import requests as req
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
url = 'https://www.fotocasa.es/indice-precio-vivienda/madrid-capital/todas-las-zonas'
html=req.get(url).text
sopa=bs(html, 'html.parser')
driver.get(url)
```

```python
boton = driver.find_element(By.CSS_SELECTOR, "div > div > div > footer > div > button.sui-AtomButton.sui-AtomButton--primary.sui-AtomButton--solid.sui-AtomButton--center")
boton.click()
```

```python
viviendas = driver.find_element(By.CLASS_NAME, "table-price_left")
columnas_viviendas = viviendas.find_elements(By.TAG_NAME, "thead")
lista = []
for elemento in columnas_viviendas:
    lista.append(elemento.text)
lista = lista[0].split()
lista = lista[0], lista[1]+" "+ lista[2], lista[3]+" "+ lista[4]+" "+lista[5]+ " " + lista[6]
list(lista)
lista
```

```python
import re
datos = viviendas.find_elements(By.TAG_NAME, "tbody")
lista2 = []
for elemento in datos:
    lista2.append(elemento.text.split('\n'))

listas_separadas = []
for cadena in lista2[0]:
    partes = re.findall(r'([\w\s-]+) (\d+\.\d+ €/m²) (\d+\.\d+ €)', cadena)
    if partes:
        listas_separadas.append(list(partes[0]))
for e in listas_separadas:
    print(e)
```

```python
columnas = ['Distrito', 'Precio medio/m²', 'Valor medio de compra']
df = pd.DataFrame(listas_separadas, columns = columnas )
df
```

```python
df = df.rename(columns={'Precio medio/m²': 'Precio medio/m2'})
```

```python
df['Precio medio/m2'] = df['Precio medio/m2'].str.replace('€/m²', '')
```

```python
df['Valor medio de compra'] = df['Valor medio de compra'].str.replace('€', '')
```

```python
df['Distrito'] = df['Distrito'].str.replace('Barrio de Salamanca', 'Salamanca')
df['Distrito'] = df['Distrito'].str.replace('Moncloa - Aravaca', 'Moncloa')
df['Distrito'] = df['Distrito'].str.replace('Fuencarral - El Pardo', 'Fuencarral')
```

```python
from funcion import quitar_tildes_y_simbolos

df['Distrito'] = df['Distrito'].apply(quitar_tildes_y_simbolos)
df
```

```python
new_record = pd.DataFrame([{'Distrito': 'Barajas', 'Precio medio/m2': '3.497', 'Valor medio de compra': '387.492'}])
df = pd.concat([df, new_record], ignore_index=True)
```
