## CREACIÓN DE TABLA CENTRAL: DISTRITOS

```python
import pandas as pd
```

CREAMOS UN DATAFRAME CON LOS 21 DISTRITOS DE MADRID

```python
distritos = ['Centro','Arganzuela','Retiro','Salamanca','Chamartín','Tetuán','Chamberí','Fuencarral','Moncloa','Latina','Carabanchel','Usera','Puente de Vallecas','Moratalaz','Ciudad Lineal','Hortaleza','Villaverde','Villa de Vallecas','Vicálvaro','San Blas','Barajas']
columnas = ['Distrito']
df = pd.DataFrame(distritos, columns = columnas )
df
```

LOS ORDENAMOS POR ORDEN ALFABÉTICO

```python
df = df.sort_values(by='Distrito')
df
```

EXPORTAMOS EL DATAFRAME Y GUARDAMOS EN UN ARCHIVO .CSV

```python
df.to_csv('distritos.csv')
```
